```python
df = sns.load_dataset("titanic")
df.head(5)
```

```python
print(df.isnull().sum())
```

```python
#total null values in the dataset
df.isnull().sum().sum()
```

```python
#percentage of null values in each column
print((df.isnull().sum()/df.shape[0])*100)
```

```python
#total percentage of null values in dataset
(df.isnull().sum().sum()/(df.shape[0]*df.shape[1])*100)
```

## HANDLING MISSING VALUES
#### METHOD 1 --> Deleting rows

```python
#deleting the column that have high percentage of null values (deck column)
df.drop(columns = ["deck"],inplace = True)
print(df.shape)    #remove the deck column
```

```python
df.dropna(axis =0,inplace = False)   #actual dataset remains same
```

#### METHOD 2 --> IMPUTATION

##### a) By Mean
  --> works well with normally distributed data

```python
df['age_mean'] = df['age'].fillna(df['age'].mean())
df[['age','age_mean']]
```

##### b) by Median
  --> works well when outliers are present in dataset

```python
df['age_median'] = df['age'].fillna(df['age'].median())
df[['age','age_mean','age_median']]
```

##### c) by mode
   --> works with categorical data

```python
#checking null value in categorical column
print(df['embarked'].isnull().sum())

#checking all unique values (contains nan)
print(df['embarked'].unique())
```

```python
#applying mode
mode_value = df[df['embarked'].notna()]['embarked'].mode()[0]
df["new_embarked"] = df['embarked'].fillna(mode_value)
df[['embarked','new_embarked']]
```

```python
#checking if still now contains any nan value
df['new_embarked'].isnull().sum()
```
