```python
# defining dataframe

import pandas as pd

data = {
    'Name': ['Aman', 'Riya', 'Karan', 'Neha', 'Arjun'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Education': ['Graduate', 'Post-Graduate', 'High School', 'Graduate', 'Post-Graduate'],
    'City': ['Delhi', 'Mumbai', 'Kolkata', 'Delhi', 'Mumbai']
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder

edu_order = [["High School","Graduate","Post-Graduate"]]
oe = OrdinalEncoder(categories = edu_order)

df["Education_Label"] = oe.fit_transform(df[["Education"]])
df
```

```python
# using map function

ord_edu = {"High School":0, "Graduate":1, "Post-Graduate":2}
df["Education_ordered"] = df["Education"].map(ord_edu)
df
```
