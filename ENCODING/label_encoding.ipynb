```python
#defining dataset
import pandas as pd

data = {
    'Name': ['Aman', 'Riya', 'Karan', 'Neha', 'Arjun'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Education': ['Graduate', 'Post-Graduate', 'High School', 'Graduate', 'Post-Graduate'],
    'City': ['Delhi', 'Mumbai', 'Kolkata', 'Delhi', 'Mumbai']
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()     #creating object
df["Gender_Label"] = le.fit_transform(df["Gender"])
df
```

```python
#for the education column

le = LabelEncoder()
df["Education_Label"] = le.fit_transform(df["Education"])
df
```
