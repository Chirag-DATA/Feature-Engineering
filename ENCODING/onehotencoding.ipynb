```python
# defining dataframe

import pandas as pd

data = {
    'Name': ['Aman', 'Riya', 'Karan', 'Neha', 'Arjun'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Education': ['Graduate', 'Post-Graduate', 'High School', 'Graduate', 'Post-Graduate'],
    'City': ['Delhi', 'Mumbai', 'Kolkata', 'Delhi', 'Mumbai']
}

df = pd.DataFrame(data)
print(df)
```

```python
#one hot encoding using pandas library
data = pd.get_dummies(df, columns = ["City"])

data.replace(True,1,inplace = True)
data.replace(False,0, inplace = True)
data
```

```python
# using sklearn library
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()    #creating object
df_ohe = encoder.fit_transform(df[["City"]]).toarray()    #create an sparse matrix

print(df_ohe)
print(encoder.get_feature_names_out(["City"]))
```

```python
#converting it into a dataframe
df_encode = pd.DataFrame(df_ohe, columns = encoder.get_feature_names_out(["City"]))
data_new = pd.concat([df, df_encode], axis = 1)
data_new
```
